```python
# Pandas is a python library, where it has several re-useable functions which is used to work with database and different types of files like exccel, csv, etc.
import pandas as pd
```

```python
#This snippet is to read the file. here. read_csv is a function exists inside pandas lib to read different types of files
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#Printing the dataset
dataset
```

```python
# selecting 'YearsExperience' column as independent variable and printing it 
independent = dataset[["YearsExperience"]]
independent
```

```python
# selecting 'Salary' column as dependent variable and printing it 
dependent = dataset[["Salary"]]
dependent
```

```python
# sklearn is a famous python lib for ML. from sklearn we are importing train_test_split which is a module inside it and it is used for spliting and validting data
# we are declaring 4 variables here, to store the training (x_train & y_train) and testing (x_test & y_test) 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent,test_size = 0.30, random_state = 0)
```

```python
#Here we are importing Linear Regression pacakge/model from sklearn to predict the values based on the input 
# LinearRegression model/pakcage is being assinged to a variable called regressor
# fit - is training the model based on teh input and output data
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
#coef_ is coefficients which determines the output based on the input and it is being assinged to varibled called weight
weight = regressor.coef_
weight
```

```python
#Storing the base value to the biase variable
biase = regressor.intercept_
biase
```

```python

#Based on the input(X_test) passing to the predict function, it will predict teh output and store it in y_pred variable
y_pred = regressor.predict(X_test)
```

```python
#metrics is a module inside sklearn package used to validate models. here we are passingn two variabels.
#inside r2_score function which is used to measure teh predictions where 0 is not good. 1 is good
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,y_pred)
```

```python
#imporing a pickle packge to store the model. filename variable -  where you plan to save the model.
import pickle
filename = "finalized_model_linear.sav"
```

```python
#dump is a function used to save the model (regressor) to a file which we have declared in the above cell
pickle.dump(regressor,open(filename, 'wb'))
```

```python
#it reads the file and passing the input 3. based on it, will predict the output. 
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb'))
result = loaded_model.predict([[3]])
```
